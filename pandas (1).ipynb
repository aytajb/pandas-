```python
df=pd.read_csv('/content/car-sales-extended-missing-data.csv')
```

```python
df.isna().sum()
```

```python
df['Make']
```

```python
df.dropna()
```

```python
df.dropna(subset='Make')
```

```python
df['Make'].mode()[0]
```

```python
df
```

```python
df['Make']=df['Make'].fillna(df['Make'].mode()[0])
df['Colour']=df['Colour'].fillna(df['Colour'].mode()[0])
df['Odometer (KM)']=df['Odometer (KM)'].fillna(df['Odometer (KM)'].median())
df['Doors']=df['Doors'].fillna(df['Doors'].median())
df['Price']=df['Price'].fillna(df['Price'].median())
```

```python
df
```

```python
df['prices_per_km']=df['Price']/df['Odometer (KM)']
df
```

```python
df.groupby(by=['Make'])['Price'].mean().plot(kind='bar')
```

```python
df2=pd.get_dummies(df['Make'])
df2
```

```python
df_alt=pd.concat([df,df2],axis=1)
df_alt
```
